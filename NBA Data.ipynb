```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import playergamelog
import pandas as pd
import matplotlib.pyplot as plt
```

```python
player_dict = players.get_players()
```

```python
franz = [player for player in player_dict if player['full_name'] == 'Franz Wagner'][0]
franz_id = franz['id']
```

```python
lamelo = [player for player in player_dict if player['full_name'] == 'LaMelo Ball'][0]
lamelo_id = lamelo['id']
```

```python
df_franz_games = playergamelog.PlayerGameLog(player_id=franz_id).get_data_frames()[0]
df_melo_games = playergamelog.PlayerGameLog(player_id=lamelo_id).get_data_frames()[0]
```

```python
df_franz_games["EFF"] = df_franz_games["PTS"] / df_franz_games["FGA"]
df_franz_games[["FGA", "PTS", "EFF"]]
```

```python
df_melo_games["EFF"] = df_melo_games["PTS"] / df_melo_games["FGA"]
df_melo_games[["FGA", "PTS", "EFF"]]
```

```python
(df_melo_games["EFF"] > 1).value_counts()
```

```python
df_franz_games["HOME_yn"] = df_franz_games["MATCHUP"].apply(lambda x: False if "@" in x.lower() else True)
df_melo_games["HOME_yn"] = df_melo_games["MATCHUP"].apply(lambda x: False if "@" in x.lower() else True)
```

```python
plt.hist(df_franz_games["EFF"])
plt.hist(df_melo_games["EFF"])
```

```python
df_franz_games["EFF>1"] = df_franz_games["EFF"].apply(lambda x: True if x>1 else False)
df_melo_games["EFF>1"] = df_melo_games["EFF"].apply(lambda x: True if x>1 else False)
```
